## Association Analysis Glossary 101

*association* and *rules* are use interchangeably in the below definitions.

**Support:** How frequent the item association appears. A high support is usually better, but a low support can be useful to find 'hidden' relationships.

**Confidence:** How reliable the item association is. For instance, A confidence of .5 means that in 50% of the cases where Diaper and Gum were purchased, the purchase also included Beer and Chips. 

**Lift:** Ratio of the confidence of the association and the expected confidence of the association. The basic rule of thumb is that a lift value close to 1 means the rules were completely independent.

## Loading data and Libraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

store = pd.read_excel('Online Retail.xlsx')
store.head()
```

```python
store.shape
```

# Data Cleaning
Removing outliers and null values

```python
store = store[store.Country != 'Unspecified']
store = store[(store.Quantity >= 0) & (store.UnitPrice >= 0)]
store.dropna(subset=['CustomerID'], inplace=True)
store['Country'].loc[store.Country == 'EIRE'] = 'Ireland'
store['Country'].loc[store.Country == 'RSA'] = 'South Africa'
```

```python
store.describe()
```

```python
store.shape
```

# EDA (Exploratory Data Analysis)

```python
# count number of purchases for each customer and select more loyal customers as those 
loyal_customers = store.groupby(['CustomerID'])['InvoiceNo'].size().reset_index()
loyal_customers = loyal_customers[loyal_customers.InvoiceNo > 14]
loyal_customers.head().sort_values('InvoiceNo', ascending=False)
```

```python
# assign the total number of purchases per customer back to the original dataframe
store['transaction'] = store.groupby(['CustomerID'])['InvoiceNo'].transform('size')
```

```python
# Top three items sold per country
top_3 = store.copy()

top_3_item = top_3.set_index('Description').groupby(['Country'])['Quantity'].nlargest(3).reset_index()
top_3_item.sort_values('Quantity', ascending=False).head(20).reset_index(drop=True)
```

## Sales Demand
Sales significantly increase in August, reaching peak in mid November and dropping dramatically in December and the first three months of the new year.

```python
# Calculating total number of transactions per year and per month.
store['Month'] = store.InvoiceDate.dt.month_name()
store['Year'] = store.InvoiceDate.dt.year
demand = store.groupby(['Year','Month'])['Quantity'].sum().reset_index()
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
demand['Month'] = pd.Categorical(demand['Month'], categories=months, ordered=True)
demand.sort_values(by=['Year', 'Month'], inplace=True)
# demand.head()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation = 45)
sns.lineplot(x=demand.Month, y=demand.Quantity, data=store[store['Year'] == 2011])
```

# Market Basket Analysis

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Remove spaces, remove null values and change InvoiceNo column to str type.
store['Description'] = store['Description'].str.strip()
# store.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
store['InvoiceNo'] = store['InvoiceNo'].astype('str')
```

```python
# UK transactions are 1 hot encoded. The Netherlands is selected to keep the data set small. It is also the second country with the most sales.
basket = (store[store['Country'] =="Netherlands"].groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
          .unstack(fill_value=0).reset_index().set_index('InvoiceNo'))
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

An itemset is considered as "frequent" if it meets a user-specified support threshold. Let's say if the support threshold is set to 0.5 (50%), a frequent itemset is defined as a set of items that occur together in at least 50% of all transactions in the database.

```python
# Generate frequent item sets using apriori algorithm
frequent_itemsets = apriori(basket_sets, min_support=0.1, use_colnames=True)
```

```python
frequent_itemsets.sort_values('support', ascending=True).head()
```

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

Consumers from The Netherlands buy different types of lunchboxes together, so we can make sure that customers that are browsing for lunchboxes are have lunchboxes in their cart, get offered more lunchboxes options.

```python
# Select only items with a lift above 6 and confidence above 0.8
rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.8)].head()
```

Combination of items purchased differ between different countries. Customers from Germany prefer to buy items other than lunchboxes, like different kinds of plasters.

```python
basket2 = (store[store['Country'] =="Germany"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

basket_sets2 = basket2.applymap(encode_units)
basket_sets2.drop('POSTAGE', inplace=True, axis=1)
frequent_itemsets2 = apriori(basket_sets2, min_support=0.05, use_colnames=True)
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=1)

rules2[(rules2['lift'] >= 4) & (rules2['confidence'] >= 0.5)]
```
